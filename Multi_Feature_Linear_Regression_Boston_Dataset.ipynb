```python
# In this case our hypothesis is Oo+O1+...o13

# X - Matrix (m x n)
# x - Vector(Single Example with n feature)

def hypothesis(x,theta):
    y_ = 0.0
    n = x.shape[0]
    for i in range(n):
        y_ += (theta[i]*x[i])
    return y_

def error(X,Y,theta):
    e = 0.0
    m = X.shape[0]
    
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        e+=(y[i]-y_)**2
    return e/m
        
    
    

def gradient(X,y,theta):
    m,n=X.shape
    
    grad = np.zeros((n,))
    # for values of j we compute gradj
    for j in range(n):
        for i in range(m):
            y_ = hypothesis(X[i],theta)
            grad[j] += (y_-y[i])*X[i][j]
    return grad/m
        
    

def gradient_descent(X,y,learning_rate=0.1,max_epochs=300):
    
    m,n = X.shape
    theta =np.zeros((n,))
    error_list = []
    for i in range(max_epochs):
        e=error(X,y,theta)
        error_list.append(e)
        
        grad = gradient(X,y,theta)
        for j in range(n):
            theta[j] = theta[j] -learning_rate*grad[j]
    return theta,error_list


```

```python
#R2 score
def r2_score(y,y_):
    num = np.sum((y-y_)**2)
    denom = np.sum((y-y.mean())**2)
    score = (1-num/denom)
    return score*100
```
